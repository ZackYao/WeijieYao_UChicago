```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df1 = pd.read_csv('/work/A-all_G-f_C-all.txt',sep='	')
df2 = pd.read_csv('/work/A-all_G-m_C-all.txt',sep='	')
nan_value = float("NaN")
df1.replace('',nan_value,inplace=True)
df1.replace(-1,nan_value,inplace=True)
df1.dropna(subset = ["country","age"], inplace=True)
df2.replace('',nan_value,inplace=True)
df2.replace(-1,nan_value,inplace=True)
df2.dropna(subset = ["country","age"], inplace=True)
df0 = pd.concat([df1,df2])
```

```
df_genre = pd.read_csv('/work/LFM-1b_UGP_noPC_allmusic.txt',sep='	')
df_genre["num_songs"] = df_genre["rnb"] + df_genre["rap"] + df_genre["electronic"] + df_genre["rock"] + \
                      + df_genre["new age"] + df_genre["classical"] + df_genre["reggae"] + df_genre["blues"] + \
                      + df_genre["country"] + df_genre["world"] + df_genre["folk"] + df_genre["easy listening"] + \
                      + df_genre["jazz"] + df_genre["vocal"] + df_genre["children's"] + df_genre["punk"] + \
                      + df_genre["alternative"] + df_genre["spoken word"] + df_genre["pop"] + df_genre["heavy metal"]
df_genre = df_genre[['user_id', 'num_songs']] 

df = df0.merge(df_genre, how='inner',on=['user_id'])
dfmale = df[(df['gender']=='m')]
dffemale = df[(df['gender']=='f')]
df_countries = df[['user_id', 'country', 'num_songs']]
df.country.value_counts().plot.pie()
```
